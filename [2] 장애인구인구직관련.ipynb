## (1) 코사인 유사도 측정 
- (텍스트 -> 벡터변환)
- 여러 컬럼으로 구성된 df의 각 행을 하나의 텍스트로 합치고, 이를 벡터로 변환
- 주어진 df에서 코사인 유사도를 측정하는 코드

```python
# 필요한 라이브러리 불러오기
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 사용자로부터 입력받을 희망 직종은 user_input으로 가정
user_input = "사무 보조"

# 직무 포스팅과 사용자 정보 추출
job_postings = df3_selected['모집직종'].fillna('')
user_info = df4_selected['희망직종'].fillna(user_input)

# Combine job_postings and user_info into a list of documents
documents = job_postings.tolist() + user_info.tolist()

# CountVectorizer를 사용하여 텍스트를 토큰 수의 행렬로 변환
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# 직무 포스팅과 사용자 정보 간의 코사인 유사도 계산
cosine_similarities = cosine_similarity(X)

# 사용자 입력에 대한 코사인 유사도 값 찾기
user_input_index = len(job_postings) + user_info.index[-1]
user_similarity = cosine_similarities[user_input_index]


# 유사도가 높은 job_posting 추천
print("Length of user_similarity:", len(user_similarity))
print("Length of df3_selected:", len(df3_selected))

print("사용자에게 추천하는 직무 포스팅:")
for i, job in enumerate(recommended_jobs, start=1):
    print(f"{i}. {job}")
```

## (2) 유저로부터 input받음 ->  job_posting 상하위 10개 출력

- (여기서 든 생각)
- 그래서 자신과 같은 장애를 가진 사람들이 어떤 직무로 구직을 했는지 알려주는 시스템도 필요함


```python
# 사용자 정보 입력 받기
user_input = input("당신의 직무 선호도를 입력하세요: ")

# 입력된 정보를 텍스트로 변환하여 documents에 추가
documents.append(user_input)

# CountVectorizer를 사용하여 텍스트를 토큰 수의 행렬로 변환
X = vectorizer.transform(documents)

# 코사인 유사도 계산
cosine_similarities = cosine_similarity(X)

# 유사도가 가장 높은 직무 포스팅 찾기
most_similar_job = documents[cosine_similarities[-1].argmax()]

# 결과 제공
print("가장 유사한 직무 포스팅: ", most_similar_job)
```

```python
# 상위 및 하위 10개의 유사도 값과 해당 직무 포스팅 찾기
top_similarities_indices = cosine_similarities[-1].argsort()[:-11:-1]
bottom_similarities_indices = cosine_similarities[-1].argsort()[:10]
top_similar_jobs = [(documents[idx], cosine_similarities[-1, idx]) for idx in top_similarities_indices]
bottom_similar_jobs = [(documents[idx], cosine_similarities[-1, idx]) for idx in bottom_similarities_indices]

# 상위 10개의 유사한 직무 포스팅 출력
print("상위 10개의 유사한 직무 포스팅:")
for i, (job, similarity) in enumerate(top_similar_jobs, start=1):
    print(f"{i}. 직무: {job}, 코사인 유사도: {similarity}")

# 하위 10개의 유사한 직무 포스팅 출력
print("\n하위 10개의 유사한 직무 포스팅:")
for i, (job, similarity) in enumerate(bottom_similar_jobs, start=1):
    print(f"{i}. 직무: {job}, 코사인 유사도: {similarity}")
```
